```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

student_df = pd.read_csv("train.csv")
#print(student_df.head())
#print(student_df.info())
#print(student_df.describe())

student_df.drop(['test preparation course','lunch'],axis = 1)
#print(student_df.head())

student_df['parental level of education'] = student_df['parental level of education'].fillna('bachelor\'s degree')
#print(student_df.head())

student_df['race/ethnicity'] = student_df['race/ethnicity'].map({
    'group A':'Asian',
    'group B':'African',
    'group C':'Afro-Asian',
    'group D':'American',
    'group E':'European'
})
print(student_df.head())

print("Sex v/s Test Prep Course")
ax = sns.countplot(x='gender',hue='test preparation course',palette='Set1',data = student_df)
ax.set(title='Sex v/s Test Preparation Course',xlabel = 'Sex',ylabel = 'Test Prep Course')
plt.show()

print("Ethnicity v/s Sex")
ax = sns.countplot(x='race/ethnicity',hue='gender',palette='Set1',data=student_df)
ax.set(title="sex v/s Ethnicity",xlabel='Race',ylabel='Gender')
plt.show()

interval = (0,40,50,60,75)
categories = ['Fail','Second Class','First Class','Distinction']
student_df['grade_math'] = pd.cut(student_df.mathscore,interval,labels=categories)
ax = sns.countplot(x='grade_math',hue='gender',palette='Set2',data = student_df)
ax.set(title='Sex v/s Math Marks',xlabel='Gender',ylabel='Count')
plt.show()

inteval = (0,40,50,60,75)
categories = ['Fail','Second Class','First Class','Distinction']
student_df['score_reading'] = pd.cut(student_df.readingscore,interval,labels=categories)
ax = sns.countplot(x='score_reading',hue='gender',palette='Set3',data=student_df)
ax.set(title='Reading score v/s Sex',xlabel='Sex',ylabel='COunt')
plt.show()
```
