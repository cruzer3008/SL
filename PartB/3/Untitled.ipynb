```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

titanic_df = pd.read_csv('train.csv')
#print(titanic_df.head(5))

titanic_df.drop(['Parch','PassengerId','Name','Ticket','SibSp'],axis = 1,inplace=True)
#print(titanic_df.head())

#print(titanic_df.describe())
#print(titanic_df.info())

titanic_df['Pclass'] = titanic_df['Pclass'].map({
    1:'Luxury Class',
    2:'Economy Class',
    3:'Lower Class'
})

#print(titanic_df.head())

titanic_df['Embarked'] = titanic_df['Embarked'].fillna("S")
print(titanic_df.head(5))

print("Survival v/s Class")
ax = sns.countplot(x='Pclass',hue='Survived',palette = 'Set1',data=titanic_df)
ax.set(title = 'Passenger Status v/s Passenger Class',xlabel = 'Passenger Class',ylabel="Total")
plt.show()

print("Survival v/s Gender")
ax = sns.countplot(x = 'Sex' , hue='Survived',palette = 'Set1',data=titanic_df)
ax.set(title="Sex v/s survival",xlabel = 'Sex',ylabel = 'Total')
plt.show()

print("Survival v/s AgeGroup")
interval = (0,18,35,60,120)
categories = ['Children','Teens','Adults','Elder']
titanic_df['Age_cats'] = pd.cut(titanic_df.Age,interval,labels=categories)
ax = sns.countplot(x='Age_cats',hue='Survived',palette='Set3',data=titanic_df)
ax.set(title='Age group v/s survived',xlabel = 'Age groups',ylabel='Count')
plt.show()
```
